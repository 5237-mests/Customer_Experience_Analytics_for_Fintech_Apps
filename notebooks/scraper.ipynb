# Scrape for CBE

```python
def fetch_reviews(app_id, bank_name, max_reviews=1000):
    all_reviews = []
    count = 0
    token = None

    while count < max_reviews:
        result, token = reviews(
            app_id,
            lang='en',
            country='us',
            count=100,
            continuation_token=token
        )
        for r in result:
            all_reviews.append({
                'review': r['content'],
                'rating': r['score'],
                'date': r['at'].strftime('%Y-%m-%d'),
                'bank': bank_name,
                'source': 'Google Play'
            })
        count = len(all_reviews)
        if not token:
            break

    return pd.DataFrame(all_reviews)
```

```python
# Fetchh reviews for CBE
app_id = "com.combanketh.mobilebanking"
bank = "Commercial Bank of Ethiopia"
cbe_df = fetch_reviews(app_id, bank)
cbe_df.to_csv("raw_cbe_bank_reviews.csv", index=False)
print("Saved reviews")
```

# Preprocessing

#### Clean cbe review data

```python
import pandas as pd

# Load cbe raw reviews from csv
cbe_df = pd.read_csv("raw_cbe_bank_reviews.csv")

cbe_df.tail()
```

```python
# check dublicates
cbe_df.duplicated().sum()
```

```python
# Drop duplicates
cbe_df.drop_duplicates(subset=["review", "bank"], inplace=True)
```

```python
# Drop empty reviews
cbe_df.dropna(subset=["review", "rating", "date"], inplace=True)
```

```python
# Format date to YYYY-MM-DD (already done in scraping)
cbe_df['date'] = pd.to_datetime(cbe_df['date']).dt.strftime('%Y-%m-%d')
```

```python
# Final structure
cbe_df = cbe_df[["review", "rating", "date", "bank", "source"]]

# Save cleaned version
cbe_df.to_csv("cleaned_cbe_bank_reviews.csv", index=False)

print(f"✅ Final dataset shape: {cbe_df.shape}")
```
